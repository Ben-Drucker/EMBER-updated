```python
import pandas as pd
import tabulate
```

```python
their_motifs = pd.read_csv('train_motifs_old.csv', header=None)[0].tolist()
```

```python
psp = pd.read_excel("../PSP.xlsx")
```

```python
all_seqs = [x.upper() for x in psp['SITE_+/-7_AA'].tolist()]
```

```python
pd.DataFrame({"seqs": all_seqs}).to_csv("all_seqs.csv", index=False)
```

```python
all_seqs_set = set(all_seqs)
```

```python
rows = []
for i, x in enumerate(their_motifs):
    if x in all_seqs_set:
        rows.append(x)
    else:
        rows.append("NULL")
```

```python
mtx = pd.read_csv("train_motifxFamMatrix.csv", header=None, sep = "\t")
```

```python
import collections

filter_dict = collections.defaultdict(list)
for i in range(len(psp)):
    filter_dict[psp.iloc[i]['SITE_+/-7_AA'].upper()].append(i)
```

```python
excel_kin = [psp.iloc[filter_dict[rows[i]]]['GENE'] if rows[i] in all_seqs_set else "NULL" for i in range(len(rows))]
excel_kin = [x.upper() if isinstance(x, str) else x for x in excel_kin]
kin_to_vec = pd.read_csv("../kin_to_vec.csv")[["Kinase", "Family"]].set_index("Kinase").to_dict()["Family"]
excel_kin_to_group = [set([kin_to_vec[y] if y in kin_to_vec else "NULL" for y in x]) for x in excel_kin]
compare_df = pd.DataFrame({"seq": rows, "mtx": mtx[0].tolist(), "excel-kin -> group": excel_kin_to_group})
```
